```python
import mediapipe as mp
import cv2
import numpy as np
import uuid
import os
```

```python
mp_drawing=mp.solutions.drawing_utils
mp_hands=mp.solutions.hands
```

# Draw Hands
<img src="https://developers.google.com/static/mediapipe/images/solutions/hand-landmarks.png"/>

```python
cap = cv2.VideoCapture(0)
with mp_hands.Hands(min_detection_confidence=0.1, min_tracking_confidence=0.1,max_num_hands=1) as hands:
# Check if the webcam is opened correctly

    while cap.isOpened():
        ret, frame = cap.read()
        
        #BGR to RGB
        image=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        
        #set flag false
        image.flags.writeable=False
        
        #detections
        results=hands.process(image)
        
        #set flag to true
        image.flags.writeable=True
        
        #RGB 2 BGR
        image=cv2.cvtColor(image,cv2.COLOR_RGB2BGR)
        
        #detections
        print(results.multi_hand_landmarks)
        
        #render results
        if results.multi_hand_landmarks:
            for num,hand in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image,hand,mp_hands.HAND_CONNECTIONS)
                
         
        cv2.imshow('SLI', image)

        if cv2.waitKey(10) & 0xFF==ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
